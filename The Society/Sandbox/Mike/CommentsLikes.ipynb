```python
import os
import matplotlib as mp
import pandas as pd
import plotly as pt
```

```python
c_pd = pd.read_csv('commentsTyler.csv') #comments csv
```

```python
tm = c_pd['CommentTime'] #Variable for comment Time Column
c_pd['CommentTime'] = pd.to_datetime(tm,unit='s', yearfirst=True) #Convert time from seconds to expanded format
c_pd = c_pd.drop(columns='CommentTimeStamp')
```

```python
c_pd['year'] = c_pd['CommentTime'].dt.year
c_pd['month'] = c_pd['CommentTime'].dt.month
c_pd['day'] = c_pd['CommentTime'].dt.day
c_pd['hour'] = c_pd['CommentTime'].dt.hour
c_pd.info()
```

```python
clean_c_pd = c_pd.drop(columns=['Attachments', 'CommentGroup', 'AttachmentTimestamp', 'UploadIP', 'Title', 'CommentAttachments'])
```
