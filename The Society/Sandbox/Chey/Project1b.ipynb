```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
posts_and_comments = os.path.join('posts_and_comments.csv')
fb_df = pd.read_csv(posts_and_comments)
fb_df
```

```python
ts = fb_df['reactions/timestamp']
fb_df['Timestamp'] = pd.to_datetime(ts,unit='s', yearfirst=True)
fb_df['year'] = fb_df['Timestamp'].dt.year
fb_df['month'] = fb_df['Timestamp'].dt.month
fb_df['day'] = fb_df['Timestamp'].dt.day
fb_df['hour'] = fb_df['Timestamp'].dt.hour
fb_df.head()
```

```python
fb_df.drop(['reactions/attachments/0/data/0/media/title'], axis=1)
fb_df.drop(['reactions/attachments/0/data/0/media/media_metadata/photo_metadata/upload_ip'], axis=1) 
fb_df.drop(['reactions/attachments/0/data/0/media/creation_timestamp'], axis=1) 
fb_df.drop(['reactions/attachments/0/data/0/media/title'], axis=1)
fb_df.drop(['reactions/attachments/0/data/0/media/uri'], axis=1)
fb_df.drop(['reactions/attachments/0/data/0/media/media_metadata/photo_metadata/upload_ip',], axis=1)
```
