```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
posts_and_comments = os.path.join('posts_and_comments.csv')
fb_df = pd.read_csv(posts_and_comments)
fb_df
```

```python
ts = fb_df['CommentTime']
fb_df = fb_df.drop(columns='CommentTimeStamp')
fb_df['CommentTime'] = pd.to_datetime(ts,unit='s', yearfirst=True)
```

```python
fb_df['year'] = fb_df['CommentTime'].dt.year
fb_df['month'] = fb_df['CommentTime'].dt.month
fb_df['day'] = fb_df['CommentTime'].dt.day
fb_df['hour'] = fb_df['CommentTime'].dt.hour
fb_df.head()
```
